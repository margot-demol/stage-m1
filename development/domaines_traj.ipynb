```python
from matplotlib import pyplot as plt
import numpy as np

uw_c=np.arange(0,5,0.1)
um_c=np.linspace(-3,4,len(uw_c))
d1=uw_c+1
d2=1-uw_c
d=np.ones_like(um_c)
d4=np.ones_like(um_c)*3
x=np.linspace(0,0.41,10)
y1=np.ones_like(x)*0.003
y2=np.ones_like(x)*0.8

fig, axes = plt.subplots(ncols=2, figsize=(12,5))

axes[0].plot(uw_c, d1, label=r'$\frac{U_m-U_w}{c}=1$',lw=2, color='darkred')
axes[0].plot(uw_c,d2, label=r'$\frac{U_m+U_w}{c}=1$',lw=2, color='lightcoral')
axes[0].set_xlabel(r'$\frac {U_w}{c}$',fontsize=18)
axes[0].set_ylabel(r'$\frac {U_m}{c}$', rotation='horizontal',fontsize=18)
axes[0].axis([0,1.5,-2,3])
axes[0].grid()

axes[0].set_title('Cas non advecté', fontsize=15)
axes[0].fill_between(uw_c, d2, d1, facecolor='cadetblue')
axes[0].fill_between(np.linspace(0,0.41,10),-0.8, 0.8, hatch='///',color='red', alpha=0.3)
axes[0].annotate('Trajectoires convergentes', (0.7,1),fontsize=13)
axes[0].annotate('Trajectoires \n oscillantes', (0.1,2), fontsize=13)
axes[0].annotate('Marées internes \n (modes graves)', (0.1,-0.6), fontsize=15, color='red')

axes[1].plot(d,um_c, label=r'$\frac{U_w}{c}=1$',lw=2, color='darkred')
axes[1].set_xlabel(r'$\frac {U_w}{c}$',fontsize=18)
axes[1].set_ylabel(r'$\frac {U_m}{c}$', rotation='horizontal',fontsize=18)
axes[1].axis([0,3,-3,4])
axes[1].grid()

axes[1].set_title('Cas advecté', fontsize=15)
axes[1].fill_betweenx(um_c,d,d4,facecolor='cadetblue')
axes[1].fill_between(np.linspace(0,3,10),-1, 1, hatch='///',color='red', alpha=0.3)
axes[1].annotate('Trajectoires convergentes', (1.1,1.5), fontsize=13)
axes[1].annotate('Trajectoires \n oscillantes', (0.1,1.5), fontsize=13)
axes[1].annotate('Ondes \n inertielles', (2,-0.5), fontsize=15, color='red')

LABEL=['[1]', '[2]', '[3]','[4]']

Um=[0.5,5,1,1]
Uw=[0.5,2,1.25,2]
Um_c=[i/2.31 for i in Um]
Uw_c=[i/2.31 for i in Uw]
axes[0].scatter(Uw_c, Um_c, color='red',label='Cas sur la figure ...')

for i in range(4):
    axes[0].annotate(LABEL[i], (Uw_c[i],Um_c[i]), fontsize=13 )

Um=[0.1,0.1,0.1,0.1]
Uw=[0.5,1.5,2.25,2.5]
Um_c=[i/2.31 for i in Um]
Uw_c=[i/2.31 for i in Uw]
axes[1].scatter(Uw_c, Um_c,color='red',label='Cas sur la figure ...' )

for i in range(4):
    axes[1].annotate(LABEL[i], (Uw_c[i],Um_c[i]),fontsize=13)

    
axes[0].legend(fontsize=13, loc='lower right')
axes[1].legend(fontsize=13)
fig.suptitle('Domaines de trajectoires', fontsize=18)
plt.tight_layout()
```

# Domaines de trajectoires pour comparaison erreur

```python
from matplotlib import pyplot as plt
import numpy as np
import stagem1.temporal_integration as sti

uw_c=np.arange(0,5,0.1)
um_c=np.linspace(-3,4,len(uw_c))
d1=uw_c+1
d2=1-uw_c
d=np.ones_like(um_c)
d4=np.ones_like(um_c)*3
x=np.linspace(0,0.41,10)
y1=np.ones_like(x)*0.003
y2=np.ones_like(x)*0.8

fig, axes = plt.subplots(ncols=2, figsize=(12,5))

axes[0].set_xlabel(r'$\frac {U_w}{c}$',fontsize=18)
axes[0].set_ylabel(r'$\frac {U_m}{c}$', rotation='horizontal',fontsize=18)
axes[0].axis([0,1.4,-1,3])
axes[0].grid()

axes[0].set_title('Cas non advecté', fontsize=15)
axes[0].fill_between(uw_c, d2, d1, facecolor='cadetblue')
axes[0].fill_between(np.linspace(0,0.41,10),-0.8, 0.8, hatch='///',color='red', alpha=0.3)
axes[0].annotate('Trajectoires convergentes', (0.3,1),fontsize=13)
axes[0].annotate('Trajectoires \n oscillantes', (0.1,2), fontsize=13)
axes[0].annotate('Marées internes \n (modes graves)', (0.1,-0.6), fontsize=15, color='red')

axes[1].set_xlabel(r'$\frac {U_w}{c}$',fontsize=18)
axes[1].set_ylabel(r'$\frac {U_m}{c}$', rotation='horizontal',fontsize=18)
axes[1].axis([0,2,-1,3])
axes[1].grid()

axes[1].set_title('Cas advecté', fontsize=15)
axes[1].fill_betweenx(um_c,d,d4,facecolor='cadetblue')
axes[1].fill_between(np.linspace(0,3,10),-1, 1, hatch='///',color='red', alpha=0.3)
axes[1].annotate('Trajectoires\n convergentes', (1.1,1.5), fontsize=13)
axes[1].annotate('Trajectoires \n oscillantes', (0.1,1.5), fontsize=13)
axes[1].annotate('Ondes \n inertielles', (0.5,-0.5), fontsize=15, color='red')

Um1=np.linspace(0,4,15)
Uw1=np.linspace(0,3,15)
W1=[x*sti.w2 for x in [0.2,0.4,1,1.5,2,2.5,3,3.5]]
K1a=list(np.linspace(0.0001,0.002,10))
K1=list(np.linspace(0.0001,0.001,10))

list_x=np.ones_like(Um1)*(0.1/2.31)
list_y=[i/2.31 for i in Um1]
axes[0].plot(list_x, list_y, label=r'$U_m$', marker='+')
axes[1].plot(list_x, list_y, label=r'$U_m$', marker='+')
for i in range(2):
    axes[i].annotate("",
            xy=(list_x[-1], list_y[-1]),
            xytext=(list_x[-5], list_y[-5]),
            arrowprops=dict(arrowstyle="-|>,head_length=1,head_width=0.3", color="blue"))

list_x=[i/2.31 for i in Uw1]
list_y=np.ones_like(Uw1)*(0.1/2.31)
axes[0].plot(list_x, list_y, label=r'$U_w$', marker='+')
axes[1].plot(list_x, list_y, label=r'$U_w$', marker='+')
for i in range(2):
    axes[i].annotate("",
            xy=(list_x[-1], list_y[-1]),
            xytext=(list_x[-5], list_y[-5]),
            arrowprops=dict(arrowstyle="-|>,head_length=1,head_width=0.3", color="orange"))
    
    
list_c=[i/sti.k2 for i in W1]
list_x=[0.1/j for j in list_c]
list_y=[0.1/j for j in list_c]
axes[0].plot(list_x, list_y,label=r'$\omega$', marker='+')
axes[1].plot(list_x, list_y,label=r'$\omega$', marker='+')


list_c=[sti.w2/i for i in K1]
list_ca=[sti.w2/i for i in K1a]
list_x=[0.1/j for j in list_c]
list_y=[0.1/j for j in list_c]
list_xa=[0.1/j for j in list_ca]
list_ya=[0.1/j for j in list_ca]
axes[0].plot(list_x, list_y, label=r'$k$', marker='+')
axes[1].plot(list_xa, list_ya,label=r'$k$', marker='+')
axes[0].annotate("",
            xy=(list_x[-1], list_y[-1]),
            xytext=(list_x[-5], list_y[-5]),
            arrowprops=dict(arrowstyle="-|>,head_length=1,head_width=0.3", color="red"))
axes[1].annotate("",
            xy=(list_xa[-1], list_ya[-1]),
            xytext=(list_xa[-5], list_ya[-5]),
            arrowprops=dict(arrowstyle="-|>,head_length=1,head_width=0.3", color="red"))


#Only one legend for all    
lines, labels = axes[0].get_legend_handles_labels()
fig.legend(lines, labels, title="Variating parameter:",bbox_to_anchor=(1,0), loc="lower right", bbox_transform=fig.transFigure, ncol=4)


plt.tight_layout(rect=[0,0.1,1,1])
```

# Schéma de convergence

```python
um=1
uw=3
phi=np.arange(0, 6, 0.1)
dx_dt=um+uw*np.cos(phi+np.pi)

plt.plot(phi, dx_dt, ls=':', label=r'$u_a(\Phi_a(t))$')
plt.plot(phi, 2.31*np.ones_like(phi), ls='-', label=r'$\frac{\omega}{k}$')
plt.xlabel(r'$\Phi_a(t)$', fontsize=15)


plt.scatter([phi[38], phi[21]], [dx_dt[38],dx_dt[21]], color='red', label=r'$u_a(\Phi_a(t))>\frac{\omega}{k}$')
plt.annotate("",
            xy=(phi[21], dx_dt[21]),
            xytext=(phi[38], dx_dt[38]),
            arrowprops=dict(arrowstyle="simple,head_length=1,head_width=1,tail_width=0.2",connectionstyle="angle3,angleA=-50,angleB=60", color="red"))


plt.scatter([phi[7], phi[20]], [dx_dt[7],dx_dt[20]], color='purple', label=r'$u_a(\Phi_a(t))<\frac{\omega}{k}$')
plt.annotate("",
            xy=(phi[20], dx_dt[20]),
            xytext=(phi[7], dx_dt[7]),
            arrowprops=dict(arrowstyle="simple,head_length=1,head_width=1,tail_width=0.2",connectionstyle="angle3,angleA=40,angleB=63", color='purple') )


plt.xticks([0],[''])
plt.yticks([-2,4,2.31],[r'$U_m-U_w$',r'$U_m+U_w$',r'$\frac{\omega}{k}$'], fontsize=20)

plt.ylabel('Vitesse', fontsize=15)
plt.title(r'Cas $U_m-U_w<c<U_m+U_w$', fontsize=18)
plt.legend(fontsize=13)
```
